```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
url = 'https://github.com/MITA-Team/Machine-Learning/raw/main/Datasets.csv'
data = pd.read_csv(url)
data
```

```python
from sklearn.preprocessing import LabelEncoder
# therapy = data['Therapy']
label_encoder = LabelEncoder()
therapy = label_encoder.fit_transform(data['Therapy'])
features = data[['Speech', 'Sensory', 'Physical', 'Social', 'Total','ASD_traits']]
```

```python
# Split data
training_size= int(len(data)*0.8)

x_train = features[:training_size]
x_test = features[training_size:]
y_train = therapy[:training_size]
y_test = therapy[training_size:]
```

```python
# Normalize data
x_train_norm = ((x_train * x_train.mean())/x_train.std()).to_numpy()
x_test_norm = ((x_test * x_test.mean())/x_test.std()).to_numpy()

# Encoded labels
y_train_encoded = tf.keras.utils.to_categorical(y_train, num_classes=12)
y_test_encoded = tf.keras.utils.to_categorical(y_test, num_classes=12)
y_test_encoded
```

```python
# Define model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128, activation='relu', input_shape=(6,)))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(12, activation='softmax'))

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
model.fit(x_train_norm, y_train_encoded, epochs=50, batch_size=32, verbose=1, validation_data=(x_test_norm,y_test_encoded))
```

# Input session

```python
# Input feature
input_speech = float(input('Masukkan Speech: '))
input_sensori = float(input('Masukkan Sensori: '))
input_physical = float(input('Masukkan Physical: '))
input_social = float(input('Masukkan Speech: '))
input_total = input_speech + input_sensori + input_physical + input_social
input_asd = float(input('ASD tidak? '))
```

```python
user_input = pd.DataFrame([[input_speech, input_sensori, input_physical, input_social, input_total, input_asd]],
                           columns=['Speech', 'Sensori', 'Physical', 'Social', 'Total', 'ASD_traits'])

# Normalisasi input
user_input_norm = (user_input - x_train.mean()) / x_train.std()
user_input_norm = user_input.to_numpy()
user_input_norm
```

```python
predictions = model.predict(user_input_norm.reshape(1, -1))

y_train_encoded = pd.DataFrame(y_train_encoded)

y_train_encoded
```

```python
# Mengambil 5 hasil prediksi tertinggi
top_n = 3
top_indices = np.argsort(predictions[0])[::-1][:top_n]
top_probabilities = predictions[0][top_indices]

# Inisialisasi objek LabelEncoder
label_encoder = LabelEncoder()

# Fit LabelEncoder pada data latih
label_encoder.fit(y_train)
label_asli = {1:'Speech 1', 2:'Speech 2', 3: 'Speech 3', 4:'Social 1', 5:'Social 2', 6:'Social 3',
              7:'Sesnory 1', 8:'Sesnory 2', 9:'Sesnory 3', 10:'Physical 1', 11:'Physical 2', 12:'Physical 3'}

# Membalikkan proses encoding pada hasil prediksi
decoded_predictions = label_encoder.inverse_transform(top_indices)

# Menampilkan hasil prediksi dalam bentuk label asli
print(f"Top {top_n} Predicted Therapies (Original Labels):")
for label, prob in zip(decoded_predictions, top_probabilities):
    original_label = label_asli[label]
    print(f"{original_label}: {prob:.4f}")
    
```
